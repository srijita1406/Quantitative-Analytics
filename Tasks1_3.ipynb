Task 1:Create a visualization to show the following in 15-minute intervals for each stock in the included data -
a.	Average spread (Ask - Bid),
b.	Average total bin volume as a percent of the full dayâ€™s volume
c.	Volume-weighted average price (VWAP)


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

def process_file(file_name):
    # Load the data
    df = pd.read_csv(file_name)
    df.columns = df.columns.str.strip()

    # Add a column 'stock' with the source file name
    df['stock'] = file_name

    # Convert the 'timestamp' column from Julian format to datetime format
    df['Timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

    # Extract Date and Time from the 'Timestamp' column
    df['Date'] = df['Timestamp'].dt.strftime('%d/%m/%Y')
    df['Time'] = df['Timestamp'].dt.strftime('%H:%M')

    # Get the indices of rows to drop (those before 09:30 and after 16:00)
    indices_to_drop = df[(df['Time'] < '09:30') | (df['Time'] > '16:00')].index

    # Drop the rows permanently
    df.drop(indices_to_drop, inplace=True)

    # Calculate the actual volumes by taking the differences between consecutive cumulative volume values
    df['Actual Volume'] = df['volume'].diff().fillna(df['volume'])  # Fill NaN with cumulative volume

    # Calculate full day's volume (sum of all actual volumes)
    full_day_volume = df['Actual Volume'].sum()

    return df, full_day_volume

def generate_result_df(processed_df):
    # Initialize an empty list to store dictionaries
    result_data = []

    # Loop through unique dates and times in processed_df
    unique_dates = processed_df['Date'].unique()
    unique_times = processed_df['Time'].unique()

    for date in unique_dates:
        for time in unique_times:
            filtered_df = processed_df[(processed_df['Date'] == date) & (processed_df['Time'] == time)]
            if not filtered_df.empty:
                highest_bid = filtered_df['bid_price'].max()
                lowest_ask = filtered_df['ask_price'].min()
                stock_name = filtered_df['stock'].iloc[0]  # Assuming stock name is the same for each date and time combination
                average_volume = filtered_df['Actual Volume'].mean()  # Taking the average if there are duplicates
                vwap = filtered_df['vwap'].iloc[0]  # VWAP at the 0th index for each time frame
                result_data.append({
                    'Timestamp': filtered_df['Timestamp'].iloc[0],
                    'Time': time,
                    'Date': date,
                    'highest_bid': highest_bid,
                    'lowest_ask': lowest_ask,
                    'stock': stock_name,
                    'average_volume': average_volume,
                    'vwap': vwap  # Include the existing VWAP column
                })

    # Create the DataFrame after the loop using the list of dictionaries
    result_df = pd.DataFrame(result_data)
    return result_df

def resample_and_plot(result_df, full_day_volume):
    # Combine 'Date' and 'Time' columns into 'Datetime' column
    result_df['Datetime'] = pd.to_datetime(result_df['Date'] + ' ' + result_df['Time'])

    # Resample into 15-minute intervals based on 'Datetime' column
    resampled_df = result_df.set_index('Datetime').resample('15T').ffill()

    # Reset the index and convert 'Datetime' back to string format
    resampled_df = resampled_df.reset_index()
    resampled_df['Datetime'] = resampled_df['Datetime'].dt.strftime('%Y-%m-%d %H:%M')

    # Add the 'average_spread' column
    resampled_df['average_spread'] = resampled_df['lowest_ask'] - resampled_df['highest_bid']

    # Divide the average volume by the full day's volume
    resampled_df['volume_percent'] = (resampled_df['average_volume'] / full_day_volume) * 100

    # Plotting the average spread vs datetime
    plt.figure(figsize=(10, 6))
    plt.plot(resampled_df['Datetime'], resampled_df['average_spread'], marker='o', linestyle='-')
    plt.title('Average Spread vs. Datetime')
    plt.xlabel('Datetime')
    plt.ylabel('Average Spread')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Plotting the average volume percent vs datetime
    plt.figure(figsize=(10, 6))
    plt.plot(resampled_df['Datetime'], resampled_df['volume_percent'], marker='o', linestyle='-')
    plt.title('Average Volume Percent vs. Datetime')
    plt.xlabel('Datetime')
    plt.ylabel('Average Volume Percent')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Plotting the VWAP vs datetime
    plt.figure(figsize=(10, 6))
    plt.plot(resampled_df['Datetime'], resampled_df['vwap'], marker='o', linestyle='-')
    plt.title('VWAP vs. Datetime')
    plt.xlabel('Datetime')
    plt.ylabel('VWAP')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    return resampled_df


file_names = ['AMZN.csv', 'BLDE.csv', 'CAL.csv','KXIN.csv','PRTS.csv','SHOO.csv','SMTC.csv','TECH.csv','TTGT.csv','XOS.csv']
for file_name in file_names:
    processed_df, full_day_volume = process_file(file_name)
    result_df = generate_result_df(processed_df)

    # Save or print the results
    print(f"Results for {file_name}:")
    print(f"Results for {file_name}:")
    print(result_df.head())
    print(f"Full Day Volume: {full_day_volume}")
    resampled_df = resample_and_plot(result_df, full_day_volume)
    print(resampled_df.head())
```

Task 2: building cross sectional linear regression model

```python
def process_file2(file_name):
    # Load the data
    df2 = pd.read_csv(file_name)
    df2.columns = df2.columns.str.strip()

    # Add a column 'stock' with the source file name
    df2['stock'] = file_name

    # Convert the 'timestamp' column from Julian format to datetime format
    df2['Timestamp'] = pd.to_datetime(df2['timestamp'], unit='ms')

    # Extract Date and Time from the 'Timestamp' column
    df2['Date'] = df2['Timestamp'].dt.strftime('%d/%m/%Y')
    df2['Time'] = df2['Timestamp'].dt.strftime('%H:%M')

    # Get the indices of rows to drop (those before 09:30 and after 16:00)
    indices_to_drop = df2[(df2['Time'] < '09:30') | (df2['Time'] > '16:00')].index

    # Drop the rows permanently
    df2.drop(indices_to_drop, inplace=True)

    return df2
```

```python
def calculate_bid_offer_imbalance(df2):
    df2['Time_unrounded'] = df2['Timestamp'].dt.strftime('%H:%M:%S')
    # Rearrange the columns to place 'Time_unrounded' immediately after 'Time'
    cols = df2.columns.tolist()
    time_idx = cols.index('Time')
    new_cols = cols[:time_idx + 1] + ['Time_unrounded'] + cols[time_idx + 1:]
    df = df2[new_cols]

    # Initialize an empty list to store dictionaries
    result_data2 = []

    # Loop through unique dates and times
    unique_dates = df2['Date'].unique()
    unique_times = df2['Time_unrounded'].unique()

    for date in unique_dates:
        for time in unique_times:
            filtered_df = df2[(df2['Date'] == date) & (df2['Time_unrounded'] == time)]
            if not filtered_df.empty:
                # Calculate Bid Offer Imbalance
                bid_offer_imbalance = (filtered_df['nbb_agg_size'].max() - filtered_df['nbo_agg_size'].min()) / \
                                      (filtered_df['nbb_agg_size'].max() + filtered_df['nbo_agg_size'].min())

                highest_bid = filtered_df['bid_price'].max()
                lowest_ask = filtered_df['ask_price'].min()
                stock_name = filtered_df['stock'].iloc[0]  # Assuming stock name is the same for each date and time combination
                trade_price = filtered_df['trade_price'].iloc[-1]  # Last trade's price
                trade_size = filtered_df['trade_size'].iloc[-1]  # Last trade's size
                result_data2.append({
                    'Timestamp': filtered_df['Timestamp'].iloc[-1],  # Assuming last trade's timestamp is used
                    'Time_unrounded': time,
                    'Date': date,
                    'highest_bid': highest_bid,
                    'lowest_ask': lowest_ask,
                    'Bid_Offer_Imbalance': bid_offer_imbalance,
                    'trade_price': trade_price,  # Adding trade_price column
                    'trade_size': trade_size,
                    'stock': stock_name,
                })

    # Create the DataFrame after the loop using the list of dictionaries
    result_df2 = pd.DataFrame(result_data2)
    # Combine 'Date' and 'Time_unrounded' columns into 'Datetime2' column
    result_df2['Datetime2'] = pd.to_datetime(result_df2['Date'] + ' ' + result_df2['Time_unrounded'])

    # Resample into 10-second intervals based on 'Datetime2' column
    resampled_df2 = result_df2.set_index('Datetime2').resample('10S').ffill()

    # Reset the index and convert 'Datetime2' back to string format
    resampled_df2 = resampled_df2.reset_index()
    resampled_df2['Datetime2'] = resampled_df2['Datetime2'].dt.strftime('%Y-%m-%d %H:%M:%S')

    return resampled_df2

def calculate_normalized_mid_price_return(resampled_df2):
    # Convert 'Datetime2' and 'Timestamp' columns to datetime
    resampled_df2['Datetime2'] = pd.to_datetime(resampled_df2['Datetime2'])
    resampled_df2['Timestamp'] = pd.to_datetime(resampled_df2['Timestamp'])

    # Calculate midpoint price at time t
    resampled_df2['midpoint_t'] = (resampled_df2['highest_bid'] + resampled_df2['lowest_ask']) / 2

    # Calculate midpoint price at time t+10 seconds
    resampled_df2['midpoint_t_plus_10'] = resampled_df2['midpoint_t'].shift(-1)  # Assuming each row represents a 10-second interval

    # Calculate the spread
    resampled_df2['spread'] = resampled_df2['lowest_ask'] - resampled_df2['highest_bid']

    # Calculate normalized mid-price return
    resampled_df2['normalized_mid_price_return'] = ((resampled_df2['midpoint_t_plus_10'] - resampled_df2['midpoint_t']) / resampled_df2['midpoint_t']) / resampled_df2['spread']

    # Discard observations where the spread is zero or negative
    resampled_df2 = resampled_df2[resampled_df2['spread'] > 0]

    # Reset the index and show the result
    resampled_df2 = resampled_df2.reset_index(drop=True)
    return resampled_df2

def calculate_trade_imbalance(resampled_df2):
    # Calculate average bid-ask price
    resampled_df2['avg_bid_ask'] = (resampled_df2['highest_bid'] + resampled_df2['lowest_ask']) / 2

    # Determine trade sign
    resampled_df2['trade_sign'] = resampled_df2.apply(lambda row: 1 if row['trade_price'] > row['avg_bid_ask'] else -1, axis=1)

    # Calculate trade size imbalance
    resampled_df2['trade_size_imbalance'] = resampled_df2['trade_sign'] * resampled_df2['trade_size']

    # Fill NaN values with 0
    resampled_df2.fillna(0, inplace=True)

    return resampled_df2
```

```python
for file_name in file_names:
    # Process the file
    processed_df2 = process_file2(file_name)
    resampled_df2 = calculate_bid_offer_imbalance(processed_df2)
    resampled_df2 = calculate_normalized_mid_price_return(resampled_df2)
    resampled_df2 = calculate_trade_imbalance(resampled_df2)

    # Print the first few rows of the processed DataFrame
    print(f"Results for {file_name}:")
    print(resampled_df2.head())
```

Task 3: Creating the model from the data above

```python
# Define a function to perform linear regression analysis
def perform_linear_regression(df):
    # Define response and predictors
    y = df['normalized_mid_price_return']
    X = df[['Bid_Offer_Imbalance', 'trade_sign', 'trade_size_imbalance']]

    # Add intercept term to predictor variables
    X = sm.add_constant(X)

    # Fit the linear regression model
    model = sm.OLS(y, X).fit()

    # Print model summary
    print(model.summary())
    # Plot the actual vs. predicted values
    predicted_values = model.predict(X)
    plt.figure(figsize=(10, 6))
    plt.scatter(y, predicted_values)
    plt.plot(y, y, color='red', linestyle='--')
    plt.title('Actual vs. Predicted Values')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.show()
```

```python
# Iterate through each file in file_names
for file_name in file_names:
    # Process the file and calculate metrics
    processed_df2 = process_file2(file_name)
    resampled_df2 = calculate_bid_offer_imbalance(processed_df2)
    resampled_df2 = calculate_normalized_mid_price_return(resampled_df2)
    resampled_df2 = calculate_trade_imbalance(resampled_df2)

    # Perform linear regression analysis
    print(f"Linear Regression Analysis for {file_name}:")
    perform_linear_regression(resampled_df2)
    print()  # Print an empty line for readability between results
```
